# **Color Space Conversion**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
# RGB to GRAY
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv.imwrite('gray_image.jpg', gray_image)
```

```python
# RGB to RGBA
rgba_image = cv.cvtColor(img, cv.COLOR_BGR2BGRA)
cv2_imshow(rgba_image)
cv.imwrite('rgba_image.png', rgba_image)
```

```python
# RGB to BGR
bgr_image = cv.cvtColor(img, cv.COLOR_RGB2BGR)
cv2_imshow(bgr_image)
cv.imwrite('bgr_image.jpg', bgr_image)
```

```python
# GRAY to RGB
gray_to_rgb = cv.cvtColor(gray_image, cv.COLOR_GRAY2RGB)
cv2_imshow(gray_to_rgb)
cv.imwrite('gray_to_rgb.jpg', gray_to_rgb)
```

```python
# GRAY to BGR
gray_to_bgr = cv.cvtColor(gray_image, cv.COLOR_GRAY2BGR)
cv2_imshow(gray_to_bgr)
cv.imwrite('gray_to_bgr.jpg', gray_to_bgr)
```
